```python
cat_col = data.select_dtypes(['object']).columns

for i in cat_col:
    t_list = data[i].unique()
    counter = 0
    for j in t_list:
        data[i][data[i] == j] = counter
        counter = counter + 1 
    data[i] = data[i].astype('int')
```

```python
data = pd.read_csv('data.csv')

le = LabelEncoder()
cat_cols = data.select_dtypes(include=['object']).columns
data[cat_cols] = data[cat_cols].apply(lambda col: le.fit_transform(col.astype(str)))
```
